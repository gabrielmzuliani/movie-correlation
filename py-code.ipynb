```python
# import libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12, 8) # Adjust the configuration of the plots we will create

# import data

df = pd.read_csv(r'D:\Projetos\Projetos-portifolio\movies-correlation\movies.csv')
```

```python
# let's look at the data

df.head()
```

```python
# let's see if there is any missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())

    print('{} - {}%'.format(col, pct_missing))
```

```python
# data types for our columns

df.dtypes
```

```python
# change data type of columns

df['budget'] = df['budget'].astype('Int64')
df['gross'] = df['gross'].astype('Int64')
df['runtime'] = df['runtime'].astype('Int64')
df['votes'] = df['votes'].astype('Int64')
```

```python
df
```

```python
# creating year releasing

df['year_release'] = df['released'].astype(str).str.extract(r'(\d{4})')

df['year_release'].astype('Int64')

df
```

```python
# ordering teh data

df = df.sort_values(by = 'gross', inplace = False, ascending = False)

df
```

```python
# drop any duplicates

df['company'].drop_duplicates().sort_values(ascending = False) # sortin the data it shows us all the values were duplicated and you could apply without the drop_duplicates to see all
```

```python
df
```

```python
# budge high correlation
# company high correlation
```

```python
# scatter plot with budge vs. gross

plt.scatter(
    x = df['budget'],
    y = df['gross']
)

plt.title('Budget vs. Gross Earnings')

plt.xlabel('Budget for Film')

plt.ylabel('Gross Earnings')

plt.show()
```

```python
df.head()
```

```python
# plot budget vs. gross using seaborn

sns.regplot(
    x = 'budget',
    y = 'gross',
    data = df,
    scatter_kws = {'color': 'red'},
    line_kws = {'color': 'blue'} 
)

plt.show()
```

```python
# let's start looking at correlation
```

```python
df.corr(method = 'pearson', numeric_only = True) # pearson, kendall, spearman
```

```python
# high correlation between budget and gross
```

```python
# correlation matrix

correlation_matrix = df.corr(method = 'pearson', numeric_only = True)

sns.heatmap(
    correlation_matrix,
    annot = True
)

plt.title('Correlation Matric for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if (df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized
```

```python
df
```

```python
# correlation matrix

correlation_matrix = df_numerized.corr(method = 'pearson', numeric_only = True)

sns.heatmap(
    correlation_matrix,
    annot = True
)

plt.title('Correlation Matric for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
df_numerized.corr()
```

```python
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
